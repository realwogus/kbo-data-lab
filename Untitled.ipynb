```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# 텍스트 입력창과 버튼 생성
name_input = widgets.Text(description='이름:')
team_input = widgets.Text(description='팀:')
button = widgets.Button(description='제출')

output = widgets.Output()

def on_button_clicked(b):
    with output:
        clear_output()
        print(f"이름: {name_input.value}, 팀: {team_input.value}")

button.on_click(on_button_clicked)

# 인터페이스 출력
display(name_input, team_input, button, output)
```

```python
import tkinter as tk
from tkinter import messagebox

# ===== 기본 창 생성 =====
root = tk.Tk()
root.title("이닝 분석기")
root.geometry("600x450")

# ===== 첫 번째 메뉴 함수 =====
def open_first_menu():
    clear_window()

    title = tk.Label(root, text="1. 이닝 시작", font=("Arial", 14))
    title.pack(pady=10)

    btn_pitcher = tk.Button(root, text="1. 투수 정보 입력", command=enter_pitcher_info)
    btn_pitcher.pack(pady=5)

    btn_batter = tk.Button(root, text="2. 타자 정보 입력", command=enter_batter_info)
    btn_batter.pack(pady=5)

    btn_exit2 = tk.Button(root, text="3. 종료", command=root.destroy)
    btn_exit2.pack(pady=5)

# ===== 기능 함수 =====
def enter_pitcher_info():
    messagebox.showinfo("투수 정보", "투수 정보를 입력하는 창입니다.")

def enter_batter_info():
    messagebox.showinfo("타자 정보", "타자 정보를 입력하는 창입니다.")

def trade_system():
    messagebox.showinfo("트레이드 시스템", "트레이드 시스템 실행 중...")

def clear_window():
    # 현재 창의 모든 위젯 삭제
    for widget in root.winfo_children():
        widget.destroy()

# ===== 초기 메뉴 구성 =====
label = tk.Label(root, text="메인 메뉴", font=("Arial", 14))
label.pack(pady=10)

btn1 = tk.Button(root, text="1. 이닝 시작", command=open_first_menu)
btn1.pack(pady=5)

btn2 = tk.Button(root, text="2. 트레이드 시스템", command=trade_system)
btn2.pack(pady=5)

btn3 = tk.Button(root, text="3. 프로그램 종료", command=root.destroy)
btn3.pack(pady=5)

# ===== 이벤트 루프 시작 =====
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class BaseballApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("기말 과제")
        self.geometry("600x450")

        self.pitcher_info = None
        self.fielder_info = None

        self.frames = {}
        for F in (MainMenu, InningMenu, InputForm):
            frame = F(parent=self, controller=self)
            self.frames[F.__name__] = frame
            frame.place(relwidth=1, relheight=1)

        self.show_frame("MainMenu")

    def show_frame(self, frame_name):
        frame = self.frames[frame_name]
        frame.tkraise()
        if hasattr(frame, 'refresh'):
            frame.refresh()

    def quit_app(self):
        self.destroy()

class MainMenu(ttk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller

        ttk.Button(self, text="1. 이닝 시작", command=lambda: controller.show_frame("InningMenu")).pack(pady=20)
        ttk.Button(self, text="2. 프로그램 종료", command=controller.quit_app).pack()

class InningMenu(ttk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.buttons = []

    def refresh(self):
        for btn in self.buttons:
            btn.destroy()
        self.buttons.clear()

        if self.controller.pitcher_info:
            btn1 = ttk.Button(self, text="1. 투수 정보 출력", command=self.show_pitcher_info)
        else:
            btn1 = ttk.Button(self, text="1. 투수 정보 입력", command=lambda: self.controller.frames["InputForm"].open(role='pitcher'))

        if self.controller.fielder_info:
            btn2 = ttk.Button(self, text="2. 타자 정보 출력", command=self.show_fielder_info)
        else:
            btn2 = ttk.Button(self, text="2. 타자 정보 입력", command=lambda: self.controller.frames["InputForm"].open(role='fielder'))

        self.buttons.extend([btn1, btn2])

        btn1.pack(pady=5)
        btn2.pack(pady=5)

        if self.controller.pitcher_info and self.controller.fielder_info:
            btn3 = ttk.Button(self, text="3. 예상 결과 보기", command=self.show_result)
            self.buttons.append(btn3)
            btn3.pack(pady=5)

            btn4 = ttk.Button(self, text="4. 프로그램 종료", command=self.controller.quit_app)
            self.buttons.append(btn4)
            btn4.pack(pady=20)
        else:
            btn3 = ttk.Button(self, text="3. 프로그램 종료", command=self.controller.quit_app)
            self.buttons.append(btn3)
            btn3.pack(pady=20)

    def show_pitcher_info(self):
        self.show_info(self.controller.pitcher_info, '투수')

    def show_fielder_info(self):
        self.show_info(self.controller.fielder_info, '타자')

    def show_info(self, info, role):
        top = tk.Toplevel(self)
        top.title(f"{role} 정보")
        text = tk.Text(top, height=5, width=40)
        text.pack(padx=10, pady=10)
        for k, v in info.items():
            text.insert(tk.END, f"{k}: {v}\n")

    def show_result(self):
        top = tk.Toplevel(self)
        top.title("예상 결과")
        ttk.Label(top, text="미완성").pack(padx=20, pady=20)

class InputForm(ttk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self.role = None

        self.label = ttk.Label(self, text="")
        self.label.pack(pady=10)

        self.name_entry = ttk.Entry(self)
        self.name_entry.pack(pady=5)
        self.name_entry.insert(0, "이름")

        self.team_entry = ttk.Entry(self)
        self.team_entry.pack(pady=5)
        self.team_entry.insert(0, "구단")

        self.save_button = ttk.Button(self, text="저장", command=self.save)
        self.save_button.pack(pady=10)

        self.back_button = ttk.Button(self, text="뒤로가기", command=lambda: controller.show_frame("InningMenu"))
        self.back_button.pack()

    def open(self, role):
        self.role = role
        self.label.config(text=f"{role} 정보 입력")
        self.name_entry.delete(0, tk.END)
        self.name_entry.insert(0, "")
        self.team_entry.delete(0, tk.END)
        self.team_entry.insert(0, "")
        self.controller.show_frame("InputForm")

    def save(self):
        name = self.name_entry.get()
        team = self.team_entry.get()
        info = {"Name": name, "Team": team}

        if self.role == 'pitcher':
            self.controller.pitcher_info = info
        elif self.role == 'fielder':
            self.controller.fielder_info = info

        self.controller.show_frame("InningMenu")

if __name__ == "__main__":
    app = BaseballApp()
    app.mainloop()
```
